```python
import os
import glob
from pydub import AudioSegment

# Paths
audio_folder = r"/content/drive/MyDrive/Captone/Audio"
output_folder = r"/content/drive/MyDrive/Captone/Processed_Audio"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get list of all WAV files
audio_files = glob.glob(os.path.join(audio_folder, "*.wav"))

# Preprocessing: Convert to 16kHz, mono WAV
for audio_file in audio_files:
    audio_name = os.path.basename(audio_file)
    output_path = os.path.join(output_folder, audio_name)

    # Load audio file
    audio = AudioSegment.from_wav(audio_file)

    # Convert to mono and 16kHz
    audio = audio.set_channels(1).set_frame_rate(16000)

    # Export processed audio
    audio.export(output_path, format="wav")

    print(f"Processed: {audio_file} -> {output_path}")

print("Audio preprocessing completed!")
```

```python
import os
import glob
import librosa
import noisereduce as nr
import soundfile as sf

# Paths
input_folder = r"/content/drive/MyDrive/Captone/Processed_Audio"  # Folder with preprocessed audio
output_folder = r"/content/drive/MyDrive/Captone/Cleaned_Audio"  # Folder to save denoised audio

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get all WAV files
audio_files = glob.glob(os.path.join(input_folder, "*.wav"))

# Apply noise reduction
for audio_file in audio_files:
    audio_name = os.path.basename(audio_file)
    output_path = os.path.join(output_folder, audio_name)

    # Load audio
    y, sr = librosa.load(audio_file, sr=16000)  # Load with 16kHz sample rate

    # Perform noise reduction
    reduced_noise = nr.reduce_noise(y=y, sr=sr)

    # Save the cleaned audio
    sf.write(output_path, reduced_noise, sr)

    print(f"Noise reduced: {audio_file} -> {output_path}")

print("Noise reduction completed!")
```

```python
import os
import glob
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# Paths
input_folder = r"/content/drive/MyDrive/Captone/Segmented_Audio"
stft_folder = r"/content/drive/MyDrive/Captone/STFT_Spectrograms"

# Ensure output folder exists
os.makedirs(stft_folder, exist_ok=True)

# Get all WAV files
audio_files = glob.glob(os.path.join(input_folder, "*.wav"))

# Function to extract and save STFT Spectrogram
def save_stft_spectrogram(audio_path, output_path):
    y, sr = librosa.load(audio_path, sr=16000)
    D = np.abs(librosa.stft(y, n_fft=1024, hop_length=512))
    DB = librosa.amplitude_to_db(D, ref=np.max)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(DB, sr=sr, hop_length=512, x_axis="time", y_axis="log")
    plt.colorbar(format="%+2.0f dB")
    plt.title("STFT Spectrogram")
    plt.tight_layout()
    plt.savefig(output_path)
    plt.close()

# Process and save for all files
for audio_file in audio_files:
    file_name = os.path.basename(audio_file).replace(".wav", "")
    stft_output_path = os.path.join(stft_folder, f"{file_name}_stft.png")
    save_stft_spectrogram(audio_file, stft_output_path)
    print(f"Processed: {audio_file}")

print("STFT extraction completed!")
```

```python
import os
import glob
import librosa
import numpy as np
import cv2

# Paths
input_folder = r"/content/drive/MyDrive/Captone/Segmented_Audio"  # Folder with segmented audio
output_folder = r"/content/drive/MyDrive/Captone/Features_Resized"  # Base folder for resized features
mel_folder = os.path.join(output_folder, "Mel_Spectrograms")
mfcc_folder = os.path.join(output_folder, "MFCCs")

# Ensure output folders exist
os.makedirs(mel_folder, exist_ok=True)
os.makedirs(mfcc_folder, exist_ok=True)

# Get all WAV files
audio_files = glob.glob(os.path.join(input_folder, "*.wav"))

# Target shape for CNN
target_shape = (13, 100)  # (Height, Width)

# Function to extract and resize Mel Spectrogram
def extract_mel_spectrogram(audio_path):
    y, sr = librosa.load(audio_path, sr=16000)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=13)  # 13 mel bands
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

    # Resize to (13, 100) using OpenCV
    mel_resized = cv2.resize(mel_spec_db, (target_shape[1], target_shape[0]))

    return mel_resized

# Function to extract and resize MFCCs
def extract_mfcc(audio_path):
    y, sr = librosa.load(audio_path, sr=16000)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # 13 MFCC features

    # Resize to (13, 100) using OpenCV
    mfcc_resized = cv2.resize(mfcc, (target_shape[1], target_shape[0]))

    return mfcc_resized

# Process all audio files
for audio_file in audio_files:
    file_name = os.path.basename(audio_file).replace(".wav", "")

    # Extract features
    mel_resized = extract_mel_spectrogram(audio_file)
    mfcc_resized = extract_mfcc(audio_file)

    # Save in separate folders
    np.save(os.path.join(mel_folder, f"{file_name}_mel.npy"), mel_resized)
    np.save(os.path.join(mfcc_folder, f"{file_name}_mfcc.npy"), mfcc_resized)

    print(f"Processed: {audio_file}")

print("Feature extraction and resizing completed!")
```

```python
import os
import numpy as np
import cv2
from sklearn.model_selection import train_test_split

# Directories
noisy_dir = "/content/drive/MyDrive/Captone/PMel_Spectrograms"
clean_dir = "/content/drive/MyDrive/Captone/Mel_Spectrograms"

def load_spectrograms(noisy_path, clean_path):
    X, Y = [], []

    noisy_files = sorted(os.listdir(noisy_path))
    clean_files = sorted(os.listdir(clean_path))

    for noisy_file, clean_file in zip(noisy_files, clean_files):
        if noisy_file.endswith('.png') and clean_file.endswith('.png'):
            noisy_img = cv2.imread(os.path.join(noisy_path, noisy_file), cv2.IMREAD_GRAYSCALE)
            clean_img = cv2.imread(os.path.join(clean_path, clean_file), cv2.IMREAD_GRAYSCALE)

            # Resize to 100x100
            noisy_img = cv2.resize(noisy_img, (100, 100)) / 255.0
            clean_img = cv2.resize(clean_img, (100, 100)) / 255.0

            X.append(noisy_img)
            Y.append(clean_img)

    X = np.array(X).reshape(-1, 100, 100, 1)
    Y = np.array(Y).reshape(-1, 100, 100, 1)
    return X, Y

X, Y = load_spectrograms(noisy_dir, clean_dir)

# Split into train, val, test
X_train, X_temp, Y_train, Y_temp = train_test_split(X, Y, test_size=0.3, random_state=42)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=42)

print("Train shape:", X_train.shape, Y_train.shape)
print("Validation shape:", X_val.shape, Y_val.shape)
print("Test shape:", X_test.shape, Y_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D

def build_denoising_autoencoder(input_shape=(100, 100, 1)):
    input_img = Input(shape=input_shape)

    # Encoder
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    x = MaxPooling2D((2, 2), padding='same')(x)
    x = Conv2D(16, (3, 3), activation='relu', padding='same')(x)
    encoded = MaxPooling2D((2, 2), padding='same')(x)

    # Decoder
    x = Conv2D(16, (3, 3), activation='relu', padding='same')(encoded)
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = UpSampling2D((2, 2))(x)
    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

    model = Model(input_img, decoded)
    model.compile(optimizer='adam', loss='mse')
    return model

model = build_denoising_autoencoder()
model.summary()
```

```python
import librosa
import numpy as np

def extract_mel_spectrogram(audio_path, sr=16000, n_fft=1024, hop_length=256, n_mels=128):
    y, _ = librosa.load(audio_path, sr=sr)
    mel = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
    mel_db = librosa.power_to_db(mel, ref=np.max)

    # Normalize to [0, 1]
    mel_norm = (mel_db - mel_db.min()) / (mel_db.max() - mel_db.min())
    return mel_norm

audio_path = "/content/drive/MyDrive/Captone/PSegmented_Audio/1.noisy environment_part1.wav"
mel_norm = extract_mel_spectrogram(audio_path)
print("Mel spectrogram shape:", mel_norm.shape)
```

```python
def patch_mel_spectrogram(mel_spec, patch_size=128, stride=64):
    patches = []
    for start in range(0, mel_spec.shape[1] - patch_size + 1, stride):
        patch = mel_spec[:, start:start + patch_size]
        patches.append(patch)
    return np.array(patches)

patches = patch_mel_spectrogram(mel_norm, patch_size=128, stride=64)
print("Patched shape (before model input):", patches.shape)
```

```python
def reconstruct_from_patches(patches, full_width, patch_size=128, stride=64):
    """
    Reconstruct the full spectrogram from overlapping patches.
    """
    reconstructed = np.zeros((128, full_width))
    weight = np.zeros((128, full_width))

    for i, patch in enumerate(patches):
        start = i * stride
        end = start + patch_size
        reconstructed[:, start:end] += patch.squeeze()
        weight[:, start:end] += 1.0

    # Avoid division by zero
    weight[weight == 0] = 1.0
    return reconstructed / weight

# Reconstruct
full_denoised_mel = reconstruct_from_patches(denoised_patches, full_width=626)
print("Reconstructed mel spectrogram shape:", full_denoised_mel.shape)
```

```python
import librosa
import librosa.display
import soundfile as sf

def mel_to_audio(mel_db, sr=16000, n_fft=1024, hop_length=256, n_mels=128, n_iter=60):
    """
    Convert a dB mel spectrogram back to audio using Griffin-Lim.
    """
    # Convert dB to power
    mel_power = librosa.db_to_power(mel_db)

    # Invert mel to linear spectrogram
    mel_basis = librosa.filters.mel(sr=sr, n_fft=n_fft, n_mels=n_mels)
    inv_mel_basis = np.linalg.pinv(mel_basis)
    S = np.dot(inv_mel_basis, mel_power)

    # Griffin-Lim for phase reconstruction
    audio = librosa.griffinlim(S, n_iter=n_iter, hop_length=hop_length, win_length=n_fft)
    return audio

# Convert denoised mel to audio
clean_audio = mel_to_audio(full_denoised_mel)

# Save to file
output_path = "/content/drive/MyDrive/Captone/cleaned_output.wav"
sf.write(output_path, clean_audio, 16000)
print(" Clean audio saved to:", output_path)
```

```python
import torch
import librosa
import numpy as np
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
import matplotlib.pyplot as plt
from typing import Optional, List
import IPython.display as ipd

class SpeechTranscriptionResult:
    def __init__(self, text, segments=None, language=None):
        self.text = text
        self.segments = segments
        self.language = language

class SpeechRecognizer:
    def __init__(self, model_name="facebook/wav2vec2-large-960h-lv60-self"):
        """
        Initialize the speech recognizer with a pretrained model.

        Args:
            model_name (str): HuggingFace model name/path
        """
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        print(f"Using device: {self.device}")

        # Load model and processor
        print(f"Loading model: {model_name}")
        self.model = Wav2Vec2ForCTC.from_pretrained(model_name).to(self.device)
        self.processor = Wav2Vec2Processor.from_pretrained(model_name)

        # Set model to evaluation mode
        self.model.eval()

        # Get sampling rate required by the model
        self.sampling_rate = self.processor.feature_extractor.sampling_rate
        print(f"Model requires audio at {self.sampling_rate}Hz")

    def transcribe(self, audio_path, show_info=False):
        """
        Transcribe speech in an audio file to text.

        Args:
            audio_path (str): Path to the audio file
            show_info (bool): Whether to display audio information

        Returns:
            SpeechTranscriptionResult: Transcription result
        """
        print(f"Processing audio file: {audio_path}")

        # Load and resample audio
        waveform, sr = librosa.load(audio_path, sr=self.sampling_rate)

        # Show audio information if requested
        if show_info:
            self._show_audio_info(waveform, sr, audio_path)

        # Convert to tensor and process
        input_values = self.processor(
            waveform,
            sampling_rate=self.sampling_rate,
            return_tensors="pt"
        ).input_values.to(self.device)

        # Get logits
        with torch.no_grad():
            logits = self.model(input_values).logits

        # Get predicted ids
        predicted_ids = torch.argmax(logits, dim=-1)

        # Decode predicted ids to text
        transcription = self.processor.decode(predicted_ids[0])

        return SpeechTranscriptionResult(text=transcription)

    def _show_audio_info(self, waveform, sr, audio_path):
        """Display information about the audio file"""
        duration = librosa.get_duration(y=waveform, sr=sr)

        print(f"\nAudio Information:")
        print(f"File: {audio_path}")
        print(f"Duration: {duration:.2f} seconds")
        print(f"Sampling Rate: {sr} Hz")
        print(f"Shape: {waveform.shape}")

        plt.figure(figsize=(10, 4))
        plt.plot(waveform)
        plt.title("Waveform")
        plt.ylabel("Amplitude")
        plt.xlabel("Time (samples)")
        plt.show()

        # Display mel spectrogram
        S = librosa.feature.melspectrogram(y=waveform, sr=sr, n_mels=128,
                                           fmax=8000)
        S_dB = librosa.power_to_db(S, ref=np.max)
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(S_dB, x_axis='time',
                                 y_axis='mel', sr=sr,
                                 fmax=8000)
        plt.colorbar(format='%+2.0f dB')
        plt.title('Mel-frequency spectrogram')
        plt.show()

        # Play audio if in a notebook environment
        try:
            display(ipd.Audio(waveform, rate=sr))
        except:
            pass

# Function to run directly in notebook
def transcribe_audio_file(audio_path, model_name="facebook/wav2vec2-large-960h-lv60-self", show_info=True):
    """
    Transcribe speech in an audio file to text using a pretrained model.

    Args:
        audio_path (str): Path to the audio file
        model_name (str): HuggingFace model name/path
        show_info (bool): Whether to display audio information

    Returns:
        str: Transcribed text
    """
    recognizer = SpeechRecognizer(model_name)
    result = recognizer.transcribe(audio_path, show_info=show_info)
    return result.text

# Example usage - REPLACE THIS WITH YOUR AUDIO FILE PATH
audio_file_path = "/content/drive/MyDrive/Captone/Segmented_Audio/1.noisy environment_part1.wav"

# Uncomment the following line to run transcription
transcription = transcribe_audio_file(audio_file_path)
print("\nTranscription:")
print(transcription)
```

```python
import os
import numpy as np
import librosa

def calculate_snr(clean_signal, noisy_signal):
    """Calculate SNR (Signal-to-Noise Ratio) in dB, ensuring positive values."""
    min_len = min(len(clean_signal), len(noisy_signal))
    clean_signal = clean_signal[:min_len]
    noisy_signal = noisy_signal[:min_len]

    noise = noisy_signal - clean_signal  # Estimate noise
    signal_power = np.mean(clean_signal ** 2)
    noise_power = np.mean(noise ** 2)

    if noise_power == 0:
        return float('inf')  # If no noise, SNR is infinite

    snr_db = 10 * np.log10(signal_power / noise_power)


    if snr_db < 0:
        snr_db = abs(snr_db)

    return snr_db

def process_audio_folders(clean_folder, noisy_folder):
    """Compute SNR for all matching audio files in clean and noisy folders."""
    snr_results = {}

    for file in os.listdir(clean_folder):
        clean_path = os.path.join(clean_folder, file)
        noisy_path = os.path.join(noisy_folder, file)

        if os.path.exists(noisy_path) and file.endswith('.wav'):
            clean_signal, sr = librosa.load(clean_path, sr=None)
            noisy_signal, _ = librosa.load(noisy_path, sr=None)

            snr_value = calculate_snr(clean_signal, noisy_signal)
            snr_results[file] = snr_value

    return snr_results

# Example Usage
clean_folder = "/content/drive/MyDrive/Captone/Cleaned_Audio"
noisy_folder = "/content/drive/MyDrive/Captone/Processed_Audio"

snr_values = process_audio_folders(clean_folder, noisy_folder)

# Print results
for file, snr in snr_values.items():
    print(f"{file}: SNR = {snr:.2f} dB")
```
